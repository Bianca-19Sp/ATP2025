```python
def ordenado(lista):
    if len(lista)<=1:
        return True
    if lista[0]>lista[1]:
        return False
    else:
        return ordenado(lista[1:])
    
print(ordenado([1,2,3]))

    
```

```python
def ordenadait(lista):
    res=True
    i=0
    while i<len(lista)-1 and res:
        if lista[i]>lista[i+1]:
            res=False
        i+=1
    return res




def ordenadafor(lista):
    res=True
    for i in range(len(lista)-2):
        if lista[i]>lista[i+1]:
            res=False
    return res


print(ordenadait([1,2,3]))
print(ordenadafor([1,2,3]))
```

```python
def existeElem(lista,elem):
    if not lista:
        return False
    if lista[0]==elem:
        return True
    else:
        return existeElem(lista[1:],elem)
    



print(existeElem([1,2,3],5))
```

```python
def existeElemit(lista,elem):
    res=False
    i=0
    while i<len(lista) and not res:
        if lista[i]==elem:
            res=True
        i+=1
    return res


print(existeElemit([1,2,3,4],7))
```

```python
def maiorDivisor(N):
    res={}
    for i in range(1,N+1):
        maiordiv=1
        for x in range(1,10):
            if i%x==0:
                if x>maiordiv:
                    maiordiv=x
                    res[i]=maiordiv
    return res

    
print(maiorDivisor(20))        
```

```python
def mmc(a,b):
    m=max(a,b)
    while True:
        if m%a==0 and m%b==0:
            return m
        m+=1




def mmcoutra(a,b):
    x,y=a,b
    while y!=0:
        x,y=y,x%y
    mdc=x
    return (a*b)//mdc



print(mmcoutra(12,16))
```

```python
def mmcsimples(a,b):
    if a>b:
        a,b=b,a
    min=a
    while min%b!=0:
        min+=a
    return min

print(mmcsimples(12,16))
```

```python
def ocupar_cama(hospital,unidade,id_cama):
    i=0
    encontrado=False
    while i<len(hospital) and not encontrado:
        nome_unidade,medicos,camas=hospital[i]
        if nome_unidade==unidade:
            j=0
            while j<len(camas) and not encontrado:
                id,ocupado=camas[j]
                if id==id_cama:
                    encontrado=True
                    if not ocupado:
                        camas[j]=(id,True)
                j+=1
        i+=1

    return hospital


print(ocupar_cama(hospital1,'Unidade de Cardiologia',102))
```
